# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
#Additional imports
load_dotenv()
from newsapi import NewsApiClient
```

```python
# Read your api key environment variable
news_api_key = os.getenv("NEWS_API_KEY")
```

```python
# Create a newsapi client
news_api = NewsApiClient(api_key=news_api_key)
```

```python
# Fetch the Bitcoin news articles
bit_news = news_api.get_everything(q="Bitcoin", language="en")
bit_news["totalResults"]
```

```python
# Fetch the Ethereum news articles
eth_news = news_api.get_everything(q="Ethereum", language="en")
eth_news["totalResults"]
```

```python
# Create the Bitcoin sentiment scores DataFrame
def df_maker(news, language):
    articles = []
    for article in news:
        try:
            sentiment = analyzer.polarity_scores(article["description"])
            title = article["title"]
            description = article["description"]
            text = article["content"]
            date = article["publishedAt"][:10]
            
            articles.append({
                "Title": title,
                "Description": description,
                "Text": text,
                "Date": date,
                "Compound": sentiment["compound"],
                "Positive": sentiment["pos"],
                "Negative": sentiment["neg"],
                "Neutral": sentiment["neu"],
            })
            
            cols = ["Compound", "Negative", "Neutral", "Positive", "Text"]
        except AttributeError:
            pass
    
    return pd.DataFrame(articles)[cols]


bit_df = df_maker(bit_news["articles"],"en")
bit_df.head()
```

```python
# Create the ethereum sentiment scores DataFrame
eth_df = df_maker(eth_news["articles"],"en")
eth_df.head()
```

```python
# Describe the Bitcoin Sentiment
bit_df.describe()
```

```python
# Describe the Ethereum Sentiment
eth_df.describe()
```

### Questions:

Q: Which coin had the highest mean positive score?

A: Bitcoin has the highest mean positive score but by a very narrow margin. Both are very close to each other.

Q: Which coin had the highest compound score?

A: Ethereum has a higher max compound score but Bitcoin has a higher mean compound score.

Q. Which coin had the highest positive score?

A: Ethereum has a higher max positive score but Bitcoin has a higher mean positive score.

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
lemmatizer = WordNetLemmatizer()
```

```python
# Expand the default stopwords list if necessary
sw_addons = {'has', 'as', 'you', 'including', 'today', 'announced', 'week', 'basically', 'also', 'file','photo', 
                 'representation','oname', 'text', 'dtype', 'object', 'has', 'representa','repr', '0',
                 'photoname', '[filename]','name','[]'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    sw = set(stopwords.words("english"))
    regex = re.compile('[^a-zA-Z ]')
    re_clean = regex.sub('', text)
    re_clean = re_clean.replace("\xa0"," ")
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    tokens = [word.lower() for word in words if word.lower() not in sw.union(sw_addons)]
    return tokens
```

```python
# Create a new tokens column for bitcoin
bit_df['Text'] = bit_df['Text'].astype('str')
bit_df
```

```python
bit_df['test'] =bit_df['Text'].apply(tokenizer)
```

```python
bit_tokens = []
for i in bit_df['Text']:
    tokenized_text = tokenizer(i)
    bit_tokens.append({'tokens':tokenized_text})

bit_tokens_df = pd.DataFrame(bit_tokens)
bit_df['tokens'] = bit_tokens_df
bit_df.head()
```

```python
# Create a new tokens column for ethereum
eth_df['Text'] = str(eth_df['Text'])

eth_tokens = []
for i in eth_df['Text']:
    tokenized_text = tokenizer(i)
    eth_tokens.append({'tokens':tokenized_text})

eth_tokens_df = pd.DataFrame(eth_tokens)
eth_df['tokens'] = eth_tokens_df
eth_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
def process_text(doc):
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', doc)
    words = word_tokenize(re_clean)
    lemmatizer = WordNetLemmatizer()
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw]
    return output

def bigram_counter(corpus): 
    big_string = " ".join(corpus)
    processed = process_text(big_string)
    bigrams = ngrams(processed, n=2)
    top_15 = dict(Counter(bigrams).most_common(15))
    return pd.DataFrame(list(top_15.items()), columns=['word','count'])

corpus = bit_df["Text"]
bigram_counter(corpus)
```

```python
# Generate the Ethereum N-grams where N=2
corpus = eth_df["Text"]
bigram_counter(corpus)
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
bit_words = []
for text in bit_df['tokens']:
    for word in text:
        bit_words.append(word)
        
token_count(bit_words)
```
